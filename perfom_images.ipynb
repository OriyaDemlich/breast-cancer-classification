```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from keras.preprocessing.image import ImageDataGenerator#, array_to_img, img_to_array, load_img
import random
from skimage import io
import matplotlib.pyplot as plt
```

# make the same number from each folder - 5000

```python
list_of_0 = os.listdir("/content/gdrive/MyDrive/mini-project/try_new_data/data/0")
list_of_1 = os.listdir("/content/gdrive/MyDrive/mini-project/try_new_data/data/1")
print(len(list_of_0))
print(len(list_of_1))

# build a dataFrame for all means
columns_ = ({'images':[], 'totals' :[], 'RED':[], 'GREEN':[], 'BLUE':[] })
```

dataFrame for all healthy img means

```python
df_RGB_0 = pd.DataFrame(columns=columns_)
current_folder = "/content/gdrive/MyDrive/mini-project/try_new_data/data/0"

control = 1

for img_file in list_of_0[:5000]:
  path_to_content = img_file
  image = io.imread(os.path.join(current_folder, path_to_content))

  total = image.ravel().mean()
  red = image[:, :, 0].ravel().mean()
  green = image[:, :, 1].ravel().mean()
  blue = image[:, :, 2].ravel().mean()
  new_row = {'images':img_file, 'totals': total, 'RED':red, 'GREEN':green, 'BLUE':blue}

  df_RGB_0 = pd.concat([df_RGB_0, pd.DataFrame([new_row])], ignore_index=True)

  print(control , end=' ')
  control += 1
```

dataFrame for all breast cancer img means

```python
df_RGB_1 = pd.DataFrame(columns=columns_)
current_folder = "/content/gdrive/MyDrive/mini-project/try_new_data/data/1"

control = 1

for img_file in list_of_1[:5000]:
  path_to_content = img_file
  image = io.imread(os.path.join(current_folder, path_to_content))

  total = image.ravel().mean()
  red = image[:, :, 0].ravel().mean()
  green = image[:, :, 1].ravel().mean()
  blue = image[:, :, 2].ravel().mean()
  new_row = {'images':img_file, 'totals': total, 'RED':red, 'GREEN':green, 'BLUE':blue}

  df_RGB_1 = pd.concat([df_RGB_1, pd.DataFrame([new_row])], ignore_index=True)

  print(control , end=' ')
  control += 1
```

print the means for every columns of means in each dataFrame

```python
print("0_FOLDER:")
print(f"totals_mean: {'%.3f' % df_RGB_0['totals'].mean()}")
print(f"RED_mean: {'%.3f' % df_RGB_0['RED'].mean()}")
print(f"GREEN_mean: {'%.3f' % df_RGB_0['GREEN'].mean()}")
print(f"BLUE_mean: {'%.3f' % df_RGB_0['BLUE'].mean()}")

print("--------------------------------------------")

print("1_FOLDER:")
print(f"totals_mean: {'%.3f' % df_RGB_1['totals'].mean()}")
print(f"RED_mean: {'%.3f' % df_RGB_1['RED'].mean()}")
print(f"GREEN_mean: {'%.3f' % df_RGB_1['GREEN'].mean()}")
print(f"BLUE_mean: {'%.3f' % df_RGB_1['BLUE'].mean()}")
```

```python
# only 5000 images
fig, axes = plt.subplots(2, 2, figsize=(21, 14))
fig.suptitle('Histograms of Means')

data = [(df_RGB_0['totals'], df_RGB_1['totals'], 'lightpink', 'royalblue', 'Total Means'),
        (df_RGB_0['RED'], df_RGB_1['RED'], 'coral', 'maroon', 'Red Means'),
        (df_RGB_0['GREEN'], df_RGB_1['GREEN'], 'darkkhaki', 'darkgreen', 'Green Means'),
        (df_RGB_0['BLUE'], df_RGB_1['BLUE'], 'mediumturquoise', 'darkslategrey', 'Blue Means')]

for i, (data_0, data_1, color_0, color_1, title) in enumerate(data):
    row = i // 2
    col = i % 2
    ax = axes[row, col]
    ax.hist(data_0, bins=100, color=color_0)
    ax.hist(data_1, bins=100, color=color_1, alpha=0.5)
    ax.legend([f'{title}_0', f'{title}_1'])
    ax.set_title(title)
    ax.set_xlabel('Means')

plt.show()
```
